```python
def count_occurrences(numbers, target):
    l = []
    for x in numbers:
        l.append(int(x))
    count=0
    for num in l:
        if num == target:
            count+=1
    return(count)

numbers = input().split()
target = int(input())

print(count_occurrences(numbers, target))
```
